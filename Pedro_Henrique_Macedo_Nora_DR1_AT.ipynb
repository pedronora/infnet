Considere o argumento abaixo:

> *Você já deve ter ouvido algum especialista dizer que as pessoas precisam dedicar, no máximo, 30% da sua renda mensal à moradia, 20% à educação e 15% ao transporte. Esses valores não devem ser totalmente desprezados, mas não podem funcionar como um norte para o orçamento doméstico de todas as famílias.* (Fonte: BTG Pactual Digital)






Crie um programa contendo uma função que, dados um valor de renda mensal total, gastos totais com moradia, gastos totais com educação e gastos totais com transporte, faça um diagnóstico da saúde financeira do usuário, com base nos valores percentuais acima expostos, informando qual é o percentual da renda mensal total comprometido por cada custo. 

Se o gasto estiver dentro do percentual recomendado, informe ainda que 

`Seus gastos estão dentro da margem recomendada.`

Caso contrário, informe:


`Portanto, idealmente, o máximo de sua renda comprometida com {tipo} deveria ser de R$ {valor_max}`
  

Onde tipo deve ser moradia, educação ou transportes e valor_max deve ser o valor equivalente ao percentual máximo de gasto com esse tipo de custo.

Exemplo de saída do programa:

```
Renda mensal total: R$ 5000
Gastos totais com moradia: R$ 1760
Gastos totais com educação: R$ 800
Gastos totais com transporte: R$ 300

Diagnóstico:
Seus gastos totais com moradia comprometem 35.2% de sua renda total. O máximo recomendado é de 30%. Portanto, idealmente, o máximo de sua renda comprometida com moradia deveria ser de R$ 1500.
Seus gastos totais com educação comprometem 16% de sua renda total. O máximo recomendado é de 20%. Seus gastos estão dentro da margem recomendada.
Seus gastos totais com transporte comprometem 6% de sua renda total. O máximo recomendado é de 15%. Seus gastos estão dentro da margem recomendada.
```

````python
def converter_moeda(valor, pos=None, simbolo='R$'):
  valor = f'{valor:,.2f}'.replace('.', 'v').replace(',', 'p')
  return simbolo + ' ' + valor.replace('p', '.').replace('v', ',')
````

````python
def diagnostico_financeiro(total, moradia, educacao, transporte):
    percentuais = {'moradia': [0.3, moradia/total], 'educação': [0.2, educacao/total], 'transporte': [0.15, transporte/total]}

    for tipo in percentuais:
        if percentuais[tipo][0] < percentuais[tipo][1]:
            print(f'Seus gastos totais com {tipo} comprometem {percentuais[tipo][1]*100:.1f}% de sua renda total. O máximo recomendado é de {percentuais[tipo][0]*100:.1f}%. Portanto, idealmente, o máximo de sua renda comprometida com moradia deveria ser de {converter_moeda(total * percentuais[tipo][0])}.')
        else:
            print(f'Seus gastos totais com {tipo} comprometem {percentuais[tipo][1]*100:.1f}% de sua renda total. O máximo recomendado é de {percentuais[tipo][0]*100:.1f}%. Seus gastos estão dentro da margem recomendada.')
````

````python
renda_total = float(input('Informe a sua renda total:\n R$ '))
despesa_moradia = float(input('Informe a sua despesa com moradia:\n R$ '))
despesa_educacao = float(input('Informe a sua despesa com educação:\n R$ '))
despesa_transporte = float(input('Informe a sua despesa com transporte:\n R$ '))

print('\nDiagnóstico:')
diagnostico_financeiro(renda_total, despesa_moradia, despesa_educacao, despesa_transporte)
````

Seja a seguinte citação:

> *Os juros compostos são a força mais poderosa do universo e a maior invenção da humanidade, porque permitem uma confiável e sistemática acumulação de riqueza*

A frase, curiosamente, é de Albert Einstein, não de algum banqueiro ou gestor de fundos de capitais.

Assim, suponha que você possui 10.000 reais iniciais, consegue aportar R$1.000 por mês e obtém um rendimento de 0,54% ao mês. Por simplicidade, suponha que você faz o aporte após o rendimento no período acontecer.

No primeiro mês, terá 10.000 reais + 0,54% deste valor = 10.054,00 reais. E, com o novo aporte, 11.054,00 reais.

No segundo mês, o valor inicial é de 11.054,00 reais. Ele rende 0,54%, totalizando 11.113,69 reais. Com o novo aporte, 12.113,69 reais, e assim sucessivamente.

Ao final de 120 meses, você terá o montante final de 187.303,05 reais.

## a. Crie um programa que ponha a hipótese de Einstein à prova. Em uma função, receba, como entrada, um montante financeiro inicial, um percentual de rendimento por período, um valor de aporte para cada período e uma quantidade de períodos.

### Exemplo de saída do programa:

```
Valor inicial: R$ 10000
Rendimento por período (%): 0.54
Aporte a cada período: R$ 1000
Total de períodos: 120 

Após 1 períodos(s), o montante será de R$11054.00.
Após 2 períodos(s), o montante será de R$12113.69.
Após 3 períodos(s), o montante será de R$13179.11.
Após 4 períodos(s), o montante será de R$14250.27.

Após 115 períodos(s), o montante será de R$177406.76.

Após 117 períodos(s), o montante será de R$181333.33.
Após 118 períodos(s), o montante será de R$183312.53.
Após 119 períodos(s), o montante será de R$185302.42.
Após 120 períodos(s), o montante será de R$187303.05.
```
  

````python
# Importação de bibliotecas necessárias para esta e a próxima questão
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
````

````python
def juros_compostos(vp, taxa, pmt, n):
    taxa = taxa/100
    valores_futuros = []
    for p in range(1, n+1):
        # p = periodo
        vf = round(vp*((1 + taxa)**p) + pmt*(((1+taxa)**(p))-1)/taxa, 2)
        print(f'Após {p} período(s), o montante será de {converter_moeda(vf)}.')
        valores_futuros.append(vf)
    return valores_futuros
````

````python
vp = float(input('Valor inicial: R$ '))
taxa = float(input('Rendimento por período (%): '))
pmt = float(input('Aporte a cada período: R$ '))
n = int(input('Total de períodos: '))
print('-='*30, '\n')

valores = juros_compostos(vp, taxa, pmt, n)
````

## b. Crie uma função que exiba um gráfico da evolução do valor acumulado, tendo como eixo das abscissas (horizontal) o número de períodos e, no eixo das ordenadas (vertical), o valor acumulado, em reais.

````python
def gerar_grafico_invest(valores_futuros, **kwargs):
    fig, ax = plt.subplots(dpi=144)
    ax.set_title('Evolução do Valor Acumulado')
    ax.set_xlabel('N. Períodos')
    ax.set_ylabel('Valor Acumulado')
    formatter = ticker.FuncFormatter(converter_moeda)
    ax.yaxis.set_major_formatter(formatter)
    ax.plot(valores_futuros, **kwargs)
    fig.tight_layout()
    return plt.show()

gerar_grafico_invest(valores, linestyle=':', color='green')
````

## c. Desenvolva uma função que, para um país, exiba o gráfico da evolução do PIB ao longo dos anos.
* A função deve receber, como entrada, o nome de um país, e exibir o gráfico para todo o período listado na tabela.
* O gráfico deve conter os valores do PIB no eixo das ordenadas (vertical) e os anos no eixo das abscissas (horizontal)

````python
def gerar_grafico_pib(pais, **kwargs):
    pibs_transposed = pibs.T
    fig, ax = plt.subplots(dpi=144)
    ax.plot(pibs_transposed[pais], **kwargs)
    ax.set_xlabel('Anos')
    ax.set_ylabel('PIB em Trilhões')
    formatter = ticker.FuncFormatter(lambda x, y: converter_moeda(x, y, '$'))
    ax.yaxis.set_major_formatter(formatter)
    plt.title(f'Evolução do PIB: {pais}')
    fig.tight_layout()
    return plt.show()

gerar_grafico_pib(pais, color='green', linestyle='--')
````
